Aim: Representation of text document in vector space model and computing similarity between two documents.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
print("VECTOR SPACE MODEL - DOCUMENT SIMILARITY\n")
document1 = input("Enter Document 1:\n")
document2 = input("\nEnter Document 2:\n")
document3 = input("\nEnter Document 3:\n")
```

```python
documents = [document1, document2, document3]
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)
similarity_matrix = cosine_similarity(tfidf_matrix)
```

```python
print("\nCosine Similarity Results:\n")

sim_doc1_doc2 = similarity_matrix[0, 1]
print(f"Similarity between Document 1 and Document 2: {sim_doc1_doc2:.4f}")
```

```python
sim_doc1_doc3 = similarity_matrix[0, 2]
print(f"Similarity between Document 1 and Document 3: {sim_doc1_doc3:.4f}")
```

```python
sim_doc2_doc3 = similarity_matrix[1, 2]
print(f"Similarity between Document 2 and Document 3: {sim_doc2_doc3:.4f}")
```

```python
import heapq

def heuristic(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])  # Manhattan distance

def astar(grid, start, goal):
    open_list = [(0, start)]
    came_from = {}
    g = {start: 0}
    closed = set()

    while open_list:
        _, current = heapq.heappop(open_list)

        if current == goal:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            return path[::-1]

        closed.add(current)

        for move in [(0,1),(1,0),(0,-1),(-1,0)]:
            neighbor = (current[0]+move[0], current[1]+move[1])

            if (0 <= neighbor[0] < len(grid) and
                0 <= neighbor[1] < len(grid[0]) and
                grid[neighbor[0]][neighbor[1]] == 0 and
                neighbor not in closed):

                temp_g = g[current] + 1

                if neighbor not in g or temp_g < g[neighbor]:
                    g[neighbor] = temp_g
                    f = temp_g + heuristic(neighbor, goal)
                    heapq.heappush(open_list, (f, neighbor))
                    came_from[neighbor] = current

    return None


# Example
grid = [
    [0,0,0],
    [1,0,1],
    [0,0,0]
]

start = (0,0)
goal = (2,2)

print("Path:", astar(grid, start, goal))
```
